```python
import random
```

```python
def get_top_x_relevant_words(word_list, x):
    for word in word_list:
        word['relevance'] = word['rank'] + word['score_of_knowledge']
    sorted_list = sorted(word_list, key=lambda x: x['relevance'])
    return sorted_list[:x]
```

```python
def learn_relevant(num=5):
    # Get the top 'num' relevant words to learn
    words_to_learn = get_top_x_relevant_words(words_db, num)
    
    # Quiz section
    for word in words_to_learn:
        print(f"Learn this word: {word['estonian']} - {word['translation']}")
        
    correct_answers = 0
    while correct_answers < num:
        correct_answers = 0  # Reset correct answers for each iteration
        for word in words_to_learn:
            # Generate choices for the quiz
            choices = [w['translation'] for w in random.sample(words_db, 4)]
            if word['translation'] not in choices:
                choices.pop()
                choices.append(word['translation'])
            random.shuffle(choices)
            
            # Quiz the user
            print(f"\nWhat is the translation for '{word['estonian']}'?")
            for i, choice in enumerate(choices, 1):
                print(f"{i}. {choice}")
            answer = input("Your choice (1-4): ")
            
            # Check answer and update score
            if choices[int(answer)-1] == word['translation']:
                word['score_of_knowledge'] += 0.1
                correct_answers += 1
                print("Correct!")
            else:
                word['score_of_knowledge'] -= 0.2
                print("Incorrect. Keep practicing!")
    
    # Generate sentences for each word
    sentences = generate_sentences(words_to_learn)
    for sentence in sentences:
        print(sentence)
```
